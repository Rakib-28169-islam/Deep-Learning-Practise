```python
# Shapes need to be in the right way
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]], dtype=torch.float32)

# torch.matmul(tensor_A, tensor_B) # (this will error) we need 3x2 and 2x3 or
tensor_A,tensor_B,tensor_A.shape,tensor_B.shape
torch.matmul(tensor_A, tensor_B.T)
```

```python
# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

```python
# prompt: give me eample of torch squeeze easy example

# Create a tensor with a dimension of size 1
x_unsqeezed = torch.randn(1,10, 1)
print(f"Original tensor: {x_unsqeezed}\n")
print(f"Original shape: {x_unsqeezed.shape}\n")
# Remove the dimension of size 1
x_squeezed = x_unsqeezed.squeeze()
print(f"Squeezed tensor: {x_squeezed}\n")
print(f"Squeezed shape: {x_squeezed.shape}")
```

```python
# prompt: give me   easy example of unsqueeze

# Add an extra dimension at a specific position
x_unsqueeze = x_squeezed.unsqueeze(dim=0) # Add a dimension of size 1 at the beginning
print(f"Unsqueeze at dim 0: {x_unsqueeze}\n")
print(f"Shape after unsqueeze at dim 0: {x_unsqueeze.shape}")

x_unsqueeze = x_squeezed.unsqueeze(dim=1) # Add a dimension of size 1 at index 1
print(f"Unsqueeze at dim 1: {x_unsqueeze}\n")
print(f"Shape after unsqueeze at dim 1: {x_unsqueeze.shape}")

# Add an extra dimension at the end (equivalent to unsqueeze(dim=-1))
x_unsqueeze = x_squeezed.unsqueeze(dim=-1)
print(f"Unsqueeze at dim -1 (end): {x_unsqueeze}\n")
print(f"Shape after unsqueeze at dim -1: {x_unsqueeze.shape}")
```
